## Data distribution

### For each column with numeric data type, how are the values distributed?

For columns with numeric data types, we calculate:

- Percentage (from 0 to 1) of missing values
- Min value
- Lower quartile value (25th percentile)
- Median value (50th percentile)
- Upper quartile value (75th percentile)
- Max value

The group will save the results to the DataFrame `num_col_info_df`, where:
- Column names are the names of the numeric columns in `df`
- Row names: “missing_ratio”, “min”, “low_quartile”, “median”, “upper_quartile”, “max”

For ease of viewing, the group rounds all values to 1 decimal place using the `.round(1)` method.

```python
num_col_info_df = None


def missing_values_count(series):
    return series.isna().sum()
    
# Define missing ratio    
def missing_ratio(series):
    percentage = series.isna().sum() / len(series) * 100
    return round(percentage, 1)

# Define lower quartile
def lower_quartile(series):
    return series.quantile(0.25).round(1)

# Define upper quartile
def upper_quartile(series):
    return series.quantile(0.75).round(1)

#Define median
def median(series):
    return series.quantile(0.75).round(1)

num_col_info_df = df.select_dtypes(include=['int64', 'float64'])

num_col_info_df = num_col_info_df.agg([missing_values_count, missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])

num_col_info_df
```

We have 5 missing values in `passenger_age` and some abnormal values (negative values) in `passenger_age`, `baggage`, `insurance_fee` and `discount_amount`

### For each column with categorical data type, how are the values distributed?

For columns with the data type `categorical`, we need to calculate:

- The percentage (from 0 to 1) of missing values
- The number of different values (not considering missing values)
- The percentage (from 0 to 100) of each value sorted by decreasing percentage (not considering missing values)

We will save the results to the DataFrame `cat_col_info_df`, where:
- Column names are the non-numeric column names in `df`
- Row names are: “missing_ratio”, “num_values”, “value_ratios”

For ease of viewing, we round all values to 1 decimal place using the `.round(1)` method.

```python
pd.set_option('display.max_colwidth', 100) # For clearly
pd.set_option('display.max_columns', None) # For clearly
```

```python
cat_col_info_df = df.select_dtypes(include='object')

# Define num values
def num_values(series):
    return len(series.dropna().unique())

# Define value ratios
def value_ratios(series):
    value_counts = series.dropna().value_counts()
    percentages = (value_counts / len(series))*100
    percentages = percentages.round(1)
    sorted_percentages = percentages.sort_values(ascending=False)
    return sorted_percentages.to_dict()
cat_col_info_df = cat_col_info_df.agg([missing_values_count, missing_ratio, num_values, value_ratios])
cat_col_info_df
```

## Feature Engineering

We can combine the dates from the departure date column and the corresponding times from the departure time column to form a complete date and time column: `departure_datetime`.

```python
df['departure_datetime'] = pd.to_datetime(
    df['departure_date'].astype(str) + ' ' + df['departure_time'].astype(str)
)

df['arrival_datetime'] = pd.to_datetime(
    df['arrival_date'].astype(str) + ' ' + df['arrival_time'].astype(str)
)
```

Also add a new column `net_revenue` to calculate the final revenue for each ticket.

```python
df['net_revenue'] = df['ticket_price'] + df['baggage'] + df['insurance_fee'] - df['discount_amount']
```

Finally, encode `passenger_id` and `airline` for easier viewing.

```python
df['passenger_id'] = 'C-' + (df['passenger_id'].astype('category').cat.codes + 1).astype(str).str.zfill(4)

df['airline'] = 'Airline-' + (df['airline'].astype('category').cat.codes + 1).astype(str).str.zfill(4)
```
